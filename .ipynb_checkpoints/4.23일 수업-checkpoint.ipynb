## 존재하지 않는 값
1. 찾아서 복구  2. 해당 행 삭제  3. 해당 열 삭제  4. 다른 값의 평균값을 입력

```python
import pandas as pd
import numpy as np
from numpy import NaN

robots = [[24, 23680], [35,NaN], [46,47350], [27,NaN]]
data = pd.DataFrame(robots, columns = ['max_speed', 'price'])
display(data)
```

```python
data.dropna(subset = ['price'], axis = 0, inplace = True) #  subset -> 목표 set 설정  //  axis=0 -> 행 삭제  //  inplace = True -> 해당 데이터 세트 자체를 수정  
display(data)
```

```python
data.dropna('columns')
```

```python
robots = [[24, 23680], [35,NaN], [46,47350], [27,NaN]]
data = pd.DataFrame(robots, columns = ['max_speed', 'price'])
mean = data['price'].mean()
display(data)
data = data.replace(NaN, mean)
display(data)
```

## Imputation 결측치 처리

```python
df.isnull() # 결측치 확인
df.isnull().sum() # 결측치 합계 출력
```

```python
cols=['Alley', 'PoolQC'] #ALLY, POOLQC 제거
df = pd.read_csv(path_house)
df = df.drop(cols, axis=1)
df.isnull().sum()
```

## zero, mean, median, min, max, freq

```python
df = pd.read_csv(path_house)
col = ["LotFrontage"]
#zero
df[col] = df[col].fillna(0) # LotFrontage의 빈 값에 0입력
# mean
df[col] = df[col].fillna(df[col].mean()) # LotFrontage의 빈 값에 나머지 값들의 평균 입력
# median
df[col] = df[col].fillna(df[col].median()) # LotFrontage의 빈 값에 나머지 값들의 중간값 입력
# min
df[col] = df[col].fillna(df[col].min()) # LotFrontage의 빈 값에 나머지 값들의 최솟값 입력
# max
df[col] = df[col].fillna(df[col].max()) # LotFrontage의 빈 값에 나머지 값들의 최댓값 입력
#freq(최빈값)
df[col] = df[col].fillna(df[col].mode()[0]) # LotFrontage의 빈 값에 나머지 값들의 최빈값 입력
```

## 시계열 데이터: 앞에 값, 뒤에 값 채우기

```python
df = pd.read_csv(path_house)
col = ["LotFrontage"]
# 앞 값으로 채우기
df[col] = df[col].fillna(method='ffill')
# 뒷 값으로 채우기
df[col] = df[col].fillna(method='bfill')
```
