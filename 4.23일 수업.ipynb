## zero, mean, median, min, max, freq

```python
df = pd.read_csv(path_house)
col = ["LotFrontage"]
#zero
df[col] = df[col].fillna(0) # LotFrontage의 빈 값에 0입력
# mean
df[col] = df[col].fillna(df[col].mean()) # LotFrontage의 빈 값에 나머지 값들의 평균 입력
# median
df[col] = df[col].fillna(df[col].median()) # LotFrontage의 빈 값에 나머지 값들의 중간값 입력
# min
df[col] = df[col].fillna(df[col].min()) # LotFrontage의 빈 값에 나머지 값들의 최솟값 입력
# max
df[col] = df[col].fillna(df[col].max()) # LotFrontage의 빈 값에 나머지 값들의 최댓값 입력
#freq(최빈값)
df[col] = df[col].fillna(df[col].mode()[0]) # LotFrontage의 빈 값에 나머지 값들의 최빈값 입력
```

## 시계열 데이터: 앞에 값, 뒤에 값 채우기

```python
df = pd.read_csv(path_house)
col = ["LotFrontage"]
# 앞 값으로 채우기
df[col] = df[col].fillna(method='ffill')
# 뒷 값으로 채우기
df[col] = df[col].fillna(method='bfill')
```

## Moving Average: 이동평균
- 이동평균 rolling(windows=x): 이전 x개의 평균
- 이동평균 피처 생성 후, 그래프 그려보기

```python
df = pd.read_csv('data/all_stocks_5yr.csv')
print(df.shape)
df_1 = df.head(300)
```

```python
df_1['ma5'] = df_1['close'].rolling(window=5).mean()
df_1['ma30'] = df_1['close'].rolling(window=30).mean()
df_1['ma60'] = df_1['close'].rolling(window=60).mean()
df_1.tail(6)
```

```python
plt.plot(df_1.index, df_1['ma5'], label = "ma5")
plt.plot(df_1.index, df_1['ma30'], label = "ma30")
plt.plot(df_1.index, df_1['ma60'], label = "ma60")
plt.plot(df_1.index, df_1['close'], label='close')
plt.legend()
plt.grid()
```
